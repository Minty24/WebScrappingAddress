```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

r = requests.get('https://www.wework.com/locations?utm_campaign=1412012108&utm_term=54321832503&utm_content=273305650071&utm_source=ads-google&utm_medium=cpc&gclid=Cj0KCQjw_7HdBRDPARIsAN_ltcIoICKjCk44jjs4C_RyJiZNXQ88jrWqllC00TKTGu8K6OGqSAN9vKUaAgkVEALw_wcB&gclsrc=aw.ds')
c=r.text

#Using BeautifulSoup find all related info.
soup = BeautifulSoup(c, 'html.parser')

#here, we extract the link of WeWork location in US, find target content of "href"
x=soup.find_all("div", class_="UScountryBucket__countryList__1pdNV")
```

```python
#add prefix website address
prefix = 'https://www.wework.com'

#Link list contains all lins from (x)
link_list = [x.get('href') for x in soup.find_all('a')]
```

```python
#the result contains all the href tag, we need to extract only text begain with /l/
link_list
```

```python
#m is the each content in link list, create a empty list and append each link to the empty (Final Link List)
final_link_list=[]
for m in link_list:
    if m and m.startswith('/l/'):
        link_final=prefix+m
        final_link_list.append(link_final)
        print(link_final)
```

```python
#creat a dictionary for all addresses 
addr_dict = {}

for link in final_link_list:
    addr = []
    url = link
    r2 = requests.get(url)
    c2=r2.text
    soup = BeautifulSoup(c2, 'html.parser')
    x=soup.find_all("div",class_=lambda x: x and x.startswith('Text-'),type="copy3")
    
    for i in x:
        unwanted = i.find('span')
        if unwanted:
            unwanted.extract()
        elif i.get_text().endswith('.'):
            pass
        else:
            print(i.get_text())
            addr.append(i.get_text())
    addr_dict[link] = addr 
```
